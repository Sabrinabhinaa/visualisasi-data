```python
# e. Subplot untuk Menampilkan Beberapa Grafik
import matplotlib.pyplot as plt

# Data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 30, 25, 35, 40, 45]

# Subplot figure
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plot for Product A
axs[0].plot(months, product_A_sales, label='Product A', color='blue', marker='o')
axs[0].set_title('Penjualan Produk A Per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

# Plot for Product B
axs[1].plot(months, product_B_sales, label='Product B', color='red', marker='x')
axs[1].set_title('Penjualan Produk B Per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

# Kelompok
plt.text(0.5, 1.08, 'KELOMPOK 12', ha='center', fontsize=14, fontweight='bold', transform=fig.transFigure)
plt.text(0.5, 1.04, 'Sabrina Ramadhani - 0110222068', ha='center', fontsize=11, transform=fig.transFigure)
plt.text(0.5, 1.01, 'Distia Fajar Familiati - 0110222163', ha='center', fontsize=11, transform=fig.transFigure)
plt.text(0.5, 0.98, 'Muhammad Fariz Zacky - 0110222227', ha='center', fontsize=11, transform=fig.transFigure)

plt.tight_layout()
plt.show()
```
